```python
import requests
from bs4 import BeautifulSoup

def get_article_links(url):
    """Fetches article links from the given URL."""
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # This will need to be adjusted based on the website's structure
    links = soup.find_all('a', href=True)
    article_links = [link['href'] for link in links if '/2023/' in link['href']]
    
    return article_links

def get_article_content(article_url):
    """Fetches the content of an article given its URL."""
    response = requests.get(article_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # This will need to be adjusted based on the website's structure
    article_text = soup.find('article').get_text(separator='\n', strip=True)
    
    return article_text
```

```python
from langchain.document_transformers import Html2TextTransformer
from IPython.display import display, HTML
html2text = Html2TextTransformer()
```

```python
from langchain.document_loaders import AsyncHtmlLoader, UnstructuredHTMLLoader, WebBaseLoader
urls = ["https://therecord.media/mgm-cyberattack-response", 
       "https://therecord.media/greater-manchester-police-ransomware-attack-britain",
       "https://therecord.media/national-cybersecurity-alert-system-idea-cisa-panel",
       "https://therecord.media/us-canada-water-commission-investigating-cyberattack"]
```

```python
loader = WebBaseLoader(urls)
docs = loader.load()
docs_transformed = html2text.transform_documents(docs)
```

```python
for i, url in enumerate(urls):
    article_title = url.split('/')[-1]
    article_title = article_title.replace('-',' ').upper()
    article_text = docs_transformed[i].page_content
#     prompt = template.format(article_title=article_title, article_text=article_text, language = 'English')
#     summary = chat_35([HumanMessage(content=prompt)])
    display (HTML(f"<h3> {article_title}</h3>" ))
#     print(summary.content)
    print (article_text)
    display (HTML("</br>"))
```
